```python
Q1 = bank.quantile(0.25)
Q3 = bank.quantile(0.75)
IQR = Q3 - Q1
```

```python
bank_out = bank[~((bank < (Q1 - 1.5 * IQR)) |(bank > (Q3 + 1.5 * IQR))).any(axis=1)] # rows without outliers
print(bank_out.shape,bank.shape)
```

```python
bank_copy_out = bank.copy(deep=True)

# Replace every outlier on the lower side by the lower whisker
for i, j in zip(np.where(bank_copy_out < Q1 - 1.5 * IQR)[0], np.where(bank_copy_out < Q1 - 1.5 * IQR)[1]): 
    
    whisker  = Q1 - 1.5 * IQR
    bank_copy_out.iloc[i,j] = whisker[j]
    
    
#Replace every outlier on the upper side by the upper whisker    
for i, j in zip(np.where(bank_copy_out > Q3 + 1.5 * IQR)[0], np.where(bank_copy_out > Q3 + 1.5 * IQR)[1]):
    
    whisker  = Q3 + 1.5 * IQR
    bank_copy_out.iloc[i,j] = whisker[j]
    
```

```python
bank_out = bank_copy_out[~((bank_copy_out < (Q1 - 1.5 * IQR)) |(bank_copy_out > (Q3 + 1.5 * IQR))).any(axis=1)] # rows without outliers
print(bank_out.shape,bank.shape)
```

### Visualizing the Outlier treated Data before Scaling

```python
plt.figure(figsize=(15,8))
sns.boxplot(data=bank_copy_out)
```

For the columns with different units of measurement, we have to scale them.
The clustering model is using the distance based model to find similarities between the rows in the table.
The aim of clustering is to group or cluster the record, that are homoneous within the group and heterogeneous between the groups.
So scaling the variables will avoid the model giving more importance for the larger magnitude columns.
We can scale and center the data, while we are unsure of the exact relationship between the attributes for the safer side.
According to our dataset, the columns are measured in different ranges (units), the column (credit_limit) is measured in 10000s, while other columns are measured in either 100s or 1000s.
So that scaling the data is necessary for the clustering in our case.

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
bank_copy_out_std = std.fit_transform(bank_copy_out)
bank_copy_out_std
```

### Forming clusters with K = 1,3,4,5 and comparing the WSS

```python
k_means = KMeans(n_clusters = 1, random_state=1)
k_means.fit(bank_copy_out_std_df)
k_means.inertia_
```

```python
k_means = KMeans(n_clusters = 3, random_state=1)
k_means.fit(bank_copy_out_std_df)
#k_means.inertia_
print("Within Cluster Sum of Squares for 3 clusters: ",k_means.inertia_)
```

```python
k_means = KMeans(n_clusters = 4, random_state=1)
k_means.fit(bank_copy_out_std_df)
#k_means.inertia_
print("Within Cluster Sum of Squares for 4 clusters: ",k_means.inertia_)
```

```python
k_means = KMeans(n_clusters = 5, random_state=1)
k_means.fit(bank_copy_out_std_df)
#k_means.inertia_
print("Within Cluster Sum of Squares for 5 clusters: ",k_means.inertia_)
```

WSS reduces as K keeps increasing

### Calculating WSS for other values of K - Elbow Method

```python
wss =[]
for i in range(1,11):
    KM = KMeans(n_clusters=i,random_state=1)
    KM.fit(bank_copy_out_std_df)
    wss.append(KM.inertia_)    
```

```python
wss
```

```python
plt.plot(range(1,11), wss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Selecting the K-Means = 2 since the curve drops much lower from K-Means= 3

### KMeans with K=2

```python
k_means = KMeans(n_clusters = 2,random_state=1)
k_means.fit(bank_copy_out_std_df)
labels = k_means.labels_
labels
```

### Cluster evaluation for 2 clusters: the silhouette score

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
# Calculating silhouette_score
#close to 1 well separated
sil2 = silhouette_score(bank_copy_out_std_df,labels,random_state=1)
print("Silhouettte Score for K=2: ",sil2)
```

### KMeans with K=3
